```python
def romanToInt(s: str) -> int:
    main_str = s
    
    dic_1 = {
        'IV': 4,
        'IX': 9,
        'XL': 40,
        'XC': 90,
        'CD': 400,
        'CM': 900
    }

    dic_2 = {        
        'I': 1,
        'V': 5,
        'X': 10,
        'L': 50,
        'C': 100,
        'D': 500,
        'M': 1000
    }


    inverse_dic = {}

    for key in dic_1:
        inverse_dic[dic_1[key]] = key
        if key in s:
            s = s.replace(key, ' ' + str(dic_1[key]))

    for key in dic_2:
        inverse_dic[dic_2[key]] = key
        if key in s:
            s = s.replace(key, ' ' + str(dic_2[key]))

    s = s.strip() # Trim method (removing all the leading and trailing spaces)

    lst_numbers = s.split(' ')

    sum_int = 0
    for i in lst_numbers:
        sum_int += int(i)

    return sum_int


def validate_roman(s:str) -> str:
    dic = {        
        'M': 1000,
        'CM': 900,
        'D': 500,
        'CD': 400,
        'C': 100,
        'XC': 90,
        'L': 50,
        'XL': 40,
        'X': 10,
        'IX': 9,
        'V': 5,
        'IV': 4,
        'I': 1,
    }
    
    num_lst = []
    sum_n = 0
    if romanToInt(s) != -1:
        sum_n = romanToInt(s)
    for key in dic:
        con = True
        while con == True:
            if (sum_n - dic[key]) >= 0:
                sum_n -= dic[key]
                num_lst.append(key)
            else:
                con = False
    
    # list to string
    roman_str = ''
    for item in num_lst:
        roman_str += item
    
    return roman_str


numbers = input('Enter Roman Numerals: ').upper()

symbol_check_error = False
new_list = list(numbers)

dic_s = {        
    'I': 1,
    'V': 5,
    'X': 10,
    'L': 50,
    'C': 100,
    'D': 500,
    'M': 1000
}

try:
    CHECK_1 = romanToInt(numbers)
    CHECK_2 = validate_roman(numbers)

    if numbers == CHECK_2:
        print('')
        print(numbers, '=', CHECK_1)
    else:
        print('-' * 80, '\n')
        print(f'=> ROMAN NUMERAL ERROR!\n\n****NOTE: The correct way of writing {romanToInt(CHECK_2)} in Roman Numeral is --->', CHECK_2)
except:
    for n in new_list:
        if (n in dic_s) == False:
            print('-' * 80, '\n')
            print(f"""=> ROMAN NUMERAL ERROR!\n\n****NOTE: For Roman numerals, only these symbols are used:\n
            Symbols\tValue\t
            I\t\t{dic_s['I']}
            V\t\t{dic_s['V']}
            X\t\t{dic_s['X']}
            L\t\t{dic_s['L']}
            C\t\t{dic_s['C']}
            D\t\t{dic_s['D']}
            M\t\t{dic_s['M']}""")
            
            symbol_check_error = True 
            break
            
    if symbol_check_error == False:
        print('=> ROMAN NUMERAL ERROR!')
```
